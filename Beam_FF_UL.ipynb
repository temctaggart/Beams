# Description:

Select Beam that will meet the strength and deflection critereia for the application.

## Variable Definitions:##

FS - Factor of Safety

P - Uniform Load

l - Length of beam

x - arbitrary distnace from the end of the beam

$\sigma_{max}=$ Yield Strength of Material

$\sigma_{allow}=\sigma_{max}/FS$

$\delta_{max}=$ maximum deflection allowed for this beam

E_mod - Modulus of Elasticity

Z - Section Modulus

# Beam Selection Steps

1. Calculate Z (use maximum stress equation, solve for Z).
1. Use calculated value of Z to look up the material size.
1. Record I for that beam.
1. Calculate I (use maximum deflection equation, solve for I).
1. Use calculated value of I to look up the material size.
1. Record Z for that beam.
1. Choose the beam that has an acceptable Z for strength and I for deflection.
1. Re-calculate using the value of I and Z for the chosen material.

```python
# Input variables

FS = 1.5

P =  10000 # lbs

l = 180 # in

x  = 0 # in

delta_max = 0.3125 # in

sigma_max = 55000 # psi

E_Mod = 29000000 # psi

sigma_allow=sigma_max/FS #psi

#print("sigma_max = ",sigma_max)

#print("sigma_allow =",sigma_allow," lb/in^2")

def Z(P,l,FS,sigma_max):
    Z=(P*l)/(12*(sigma_max/FS))
    return Z

def I(P,l,E,delta_max):
    I=(P*(l**3))/(384*E_Mod*delta_max)
    return I

z = Z(P,l,FS,sigma_max)

print(f'Z = {z:.3f}')

print(f'Z = {Z(P,l,FS,sigma_max):.4f} in^3')

print(f'I = {I(P,l,E_Mod,delta_max):.4f} in^4')
```
